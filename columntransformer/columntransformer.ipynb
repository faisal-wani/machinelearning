```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
df.isnull().sum()

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

```python
X_train
```

## Mentos Zindagi

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
 transformer.fit_transform(X_train)
```

```python
transformer.transform(X_test)
```

```python

  
# Define ColumnTransformer
  # Keep 'age' as is

# Fit-transform on training data
X_train_transformed = transformer.fit_transform(X_train)

# Transform test data
X_test_transformed = transformer.transform(X_test)

# âœ… Correct way to extract OneHotEncoder column names
onehot_encoder = transformer.named_transformers_['tnf3']
onehot_columns = onehot_encoder.get_feature_names_out()  # Extract all names at once

# Construct final column names dynamically
column_names = ['fever', 'cough_encoded'] + list(onehot_columns) + ['age']  

# Convert transformed data into DataFrame
X_train_df = pd.DataFrame(X_train_transformed, columns=column_names)
X_test_df = pd.DataFrame(X_test_transformed, columns=column_names)

# Print transformed training and testing data
print("Transformed X_train:")
print(X_train_df)

print("\nTransformed X_test:")
print(X_test_df)
```
